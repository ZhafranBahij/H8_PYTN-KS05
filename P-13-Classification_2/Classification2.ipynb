## Naive Bayes Classifier

```python
weather = [
    'Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy',
    'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast',
    'Overcast', 'Rainy'
]
temp = [
    'Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild',
    'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'
]
play = [
    'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 
    'Yes', 'Yes', 'Yes', 'Yes', 'No'
]
```

```python
# Import LabelEncoder
from sklearn import preprocessing

# buat label endocer
le = preprocessing.LabelEncoder()

# konversi string label menjadi nomor
weather_encoded = le.fit_transform(weather)
print(weather_encoded)
```

```python
temp_encoded = le.fit_transform(temp)
label = le.fit_transform(play)
print(temp_encoded, label)
```

```python
# Mengkombinasikan weather dan temp menjadi satu list tuple
features = list(zip(weather_encoded, temp_encoded))
print(features)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB().fit(features,label)

predicted = model.predict([[0,2]])
print('Predicted Value:', predicted)
```

## Decision Tree Classifier

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv(
    './diabetes.csv',
    header=None,
    names=col_names
)
```

```python
pima.info()
```

```python
numer = col_names
for col in numer:
    pima[col] = pd.to_numeric(pima[col], errors='coerce')
```

```python
pima
```

```python
pima.dropna(inplace=True)
```

```python
pima
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
```

```python
X = pima[feature_cols] # Features
y = pima.label # Target variable
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split

#split dataset
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.3,
                                                    random_state=1
                                                   )
```

```python
clf = DecisionTreeClassifier().fit(X_train, y_train)
y_predict = clf.predict(X_test)
```

```python
from sklearn import metrics

# Model akurasi
print('Accuracy:', metrics.accuracy_score(y_test, y_predict))
```

```python
import sklearn.tree as tree
import pydotplus
from six import StringIO 
from IPython.display import Image
import graphviz
dot_data = StringIO()
tree.export_graphviz(
    clf,
    out_file=dot_data,
    class_names=['0','1'], # the target names. 
    feature_names=feature_cols, # the feature names. 
    filled=True, # Whether to fill in the boxes with colours. 
    rounded=True, # Whether to round the corners of the boxes. 
    special_characters=True)
# Image(pydotplus.graph_from_dot_data(dot_data.getvalue()))

graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```

```python
tree.plot_tree(clf)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                      class_names=['0','1'], # the target names. 
                      feature_names=feature_cols, # the feature names. 
                      filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
Image() 
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3).fit(X_train, y_train)
y_predict = clf.predict(X_test)

from sklearn import metrics

# Model akurasi
print('Accuracy:', metrics.accuracy_score(y_test, y_predict))
```

## Random Forest

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
print(iris.target_names) # Label species
print(iris.feature_names) # Features
```

```python
iris.data
```

```python
print(iris.data[-5:])
print(iris.target[-5:])
```

```python
import pandas as pd
data=pd.DataFrame({ 
    'sepal length':iris.data[:,0],
    'sepal width':iris.data[:,1],
    'petal length':iris.data[:,2],
    'petal width':iris.data[:,3],
    'species':iris.target }) 
data.head()
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split

#split dataset
X_train, X_test, y_train, y_test = train_test_split(data[['sepal length',
                                                          'sepal width',
                                                          'petal length',
                                                          'petal width'
                                                         ]],
                                                    data[['species']], 
                                                    test_size=0.3,
                                                    random_state=1
                                                   )
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100).fit(X_train, y_train)
y_predict = clf.predict(X_test)


from sklearn import metrics

# Model akurasi
print('Accuracy:', metrics.accuracy_score(y_test, y_predict))
```

## SVM

```python
from sklearn import datasets
cancer = datasets.load_breast_cancer()
```

```python
cancer.data[0:5]
```

```python
print(cancer.target)
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split

#split dataset
X_train, X_test, y_train, y_test = train_test_split(cancer.data,
                                                    cancer.target, 
                                                    test_size=0.3,
                                                    random_state=1
                                                   )
```

```python
from sklearn import svm

clf = svm.SVC(kernel='linear').fit(X_train, y_train)
y_predict = clf.predict(X_test)

from sklearn import metrics

# Model akurasi
print('Accuracy:', metrics.accuracy_score(y_test, y_predict))
```
