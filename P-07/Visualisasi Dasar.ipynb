```python
# Pie Chart

df_continents = df_can.groupby('Continent').sum()
```

```python
df_continents
```

```python
df_continents['Total'].plot(
    kind='pie', figsize=(5,6)
)
plt.axis('equal')
plt.show()
```

```python
df_continents['Total'].plot(
    kind='pie', figsize=(5,6),
    autopct='%1.3f%%'
)
plt.axis('equal')
plt.show()
```

```python
df_continents['Total'].plot(
    kind='pie', figsize=(5,6),
    autopct='%1.3f%%', startangle=113
)
plt.axis('equal')
plt.show()
```

```python
df_continents['Total'].plot(
    kind='pie', figsize=(5,6),
    autopct='%1.3f%%', startangle=-113
)
plt.axis('equal')
plt.show()
```

```python
df_continents['Total'].plot(
    kind='pie', figsize=(5,6),
    autopct='%1.3f%%', startangle=-113,
    shadow=True
)
plt.axis('equal')
plt.show()
```

```python
df_continents['Total'].plot(
    kind='pie', figsize=(15,6),
    autopct='%1.3f%%', startangle=-113,
    shadow=True, pctdistance = 1.12, labels=None
)

plt.title('Data Imigran', y=1.12)

plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')
```

```python
explode_list = [0.1, 0, 0, 0, 0, 0.1]

df_continents['Total'].plot(
    kind='pie', figsize=(15,6),
    autopct='%1.3f%%', startangle=-113,
    shadow=True, pctdistance = 1.12,
    labels=None, explode = explode_list
)

plt.title('Data Imigran', y=1.12)

plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')
```
